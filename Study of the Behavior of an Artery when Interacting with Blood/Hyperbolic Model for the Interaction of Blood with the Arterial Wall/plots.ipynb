```python
from main_Newmark import plot_p1,Newmark
from main_LeapFrog import LeapFrog
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
#plt.rcParams["figure.figsize"] = (10,6)

from scipy import interpolate


from mpl_toolkits import mplot3d
from matplotlib import cm
```

```python
k = 1

nm = Newmark(problem=1)

u_out,x,t = nm.calculate_uk_NW(k)

for j in range(10):
    plt.plot(x,u_out[j,:], label='t='+str(int(t[j]*10)/10));
plt.xlabel('x [m]')
plt.ylabel('u [m]')
plt.legend()
fig = plt.gcf()
fig.set_size_inches(6, 4)

p = nm.p_convergence(k)
print(p)
```

```python
k=2

lp = LeapFrog(problem=2)

u_out,x,t = lp.calculate_uk_LF(k)

fig, ax = plt.subplots(subplot_kw={'projection': '3d'})

X,T = np.meshgrid(x,t)
surf = ax.plot_surface(X,T,u_out, cmap = cm.viridis)
ax.set_xlabel('$x [m]$')
ax.set_ylabel('$t [s]$')
ax.set_zlabel('$u [m]$')
#ax.zaxis.set_rotate_label(False) 
ax.view_init(25,20)
ax.zaxis.set_rotate_label(False) 
ax.set_zlabel('$u [m]$', rotation = 90)

plt.show();
```

```python
k=2

lp = LeapFrog(problem=2)

u_out,x,t = lp.calculate_uk_LF(k)


t = np.linspace(0.1,1,10)
n_p = int(400/(10))
u_j = np.zeros((10,len(x)))

for j in range(1,11):
    u_j[j-1,:] = u_out[j*n_p,:]


for j in range(10):
    plt.plot(x,u_j[j,:], label='t='+str(int(t[j]*10)/10));
plt.xlabel('x [m]')
plt.ylabel('u [m]')
plt.legend(loc='upper left', prop={'size': 6})
fig = plt.gcf()
fig.set_size_inches(6, 4)
```
