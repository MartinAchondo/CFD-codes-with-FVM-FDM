```python
from main_4_1 import Theta_Method
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
#plt.rcParams["figure.figsize"] = (10,6)

from scipy import interpolate
```

### Transitoria

$\beta = \sqrt{\alpha}$

```python
th = Theta_Method()

th.set_params()
th.beta = np.sqrt(th.alpha)

tf = 2.5*10**-3
h = 10**-4

theta = 0.5
y1_cn,y2_cn,t_cn = th.solve_system(theta,tf,h)
theta = 1
y1_be,y2_be,t_be = th.solve_system(theta,tf,h)

print(th.eigen_values())
```

```python
plt.plot(y1_be,y2_be, label="BE", linewidth=1, marker='o',color='r', fillstyle='none')
plt.plot(y1_cn,y2_cn, label="CN", linewidth=1, marker='o',color='b', fillstyle='none')
plt.legend()
plt.grid()
plt.xlabel('y [m]')
plt.ylabel('dy/dt [ms-1]')
#plt.title("Residuales en N-S")

fig = plt.gcf()
fig.set_size_inches(6, 4)
```

```python
plt.plot(t_be,y1_be, label="BE", linewidth=1, marker='o',color='r', fillstyle='none')
plt.plot(t_cn,y1_cn, label="CN", linewidth=1, marker='o',color='b', fillstyle='none')
plt.legend()
plt.grid()
plt.xlabel('t [s]')
plt.ylabel('y [m]')
#plt.title("Residuales en N-S")

fig = plt.gcf()
fig.set_size_inches(6, 4)
```

```python
plt.plot(t_be,y2_be, label="BE", linewidth=1, marker='o',color='r', fillstyle='none')
plt.plot(t_cn,y2_cn, label="CN", linewidth=1, marker='o',color='b', fillstyle='none')
plt.legend()
plt.grid()
plt.xlabel('t [s]')
plt.ylabel('dy/dt [ms-1]')
#plt.title("Residuales en N-S")

fig = plt.gcf()
fig.set_size_inches(6, 4)
```
